### 데이터 셋


```python
from sklearn import datasets
from sklearn.model_selection import cross_val_score
```

```python
Data.columns
```

```python
features = Data.columns.difference(['result'])
```

```python
features = ['A153','A166','A196', 'A2', 'A239', 'A252', 'A26', 'A285', 'A304', 'A318',
           'A333','A344', 'A348', 'A359', 'A40', 'A409', 'A455', 'A475', 'A520', 
           'A523','A59', 'A65','A99'  ]
```

```python
X = Data[features]
y = Data.result
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, test_size=0.3)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Decsion Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
DT_model = DecisionTreeClassifier(max_depth=7)
```

```python
#Cross validataion
scores = cross_val_score(DT_model, X_train, y_train, cv=4)
print("Corss validation scores :{}".format(scores))
print("Corss validation means :{:.3f}".format(scores.mean()))
```

```python
DT_model.fit(X_train, y_train)
```

```python
prediction_dt = DT_model.predict(X_test)
prediction_dt
```

```python
list(zip(X.columns, DT_model.feature_importances_))
```

```python
X_train.columns
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, prediction_dt)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score


def get_clf_eval(y_test, y_pred):
    confusion = confusion_matrix(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    F1 = f1_score(y_test, y_pred)
    print('Confusion Matrix\n', confusion)
    print('\nAccuracy: {:.4f}'.format(accuracy))
    print('Precision(정밀도): {:.4f}'.format(precision))
    print('Recall(재현율): {:.4f}'.format(recall))
    print('F1: {:.4f}'.format(F1))
```

```python
get_clf_eval(y_test, prediction_dt)
```

### RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

RF_model = RandomForestClassifier(n_estimators = 1000, random_state=13, max_depth=7)

param_grid = {
    'max_depth': [3, 5, 6, 8],
    'min_samples_split': [0.01, 0.05, 0.1],
    'max_features': ['auto', 'sqrt'],
}

```

```python
RF_model.fit(X_train, y_train)
```

```python
gridsearch_rf = GridSearchCV(estimator=RF_model, param_grid=param_grid, scoring='accuracy', cv=5, n_jobs=-1)
gridsearch_rf.fit(X_train, y_train)
```

```python
print('best params', gridsearch_rf.best_params_)
print('best score', gridsearch_rf.best_score_) 
```

```python
#Cross validataion
scores = cross_val_score(gridsearch_rf, X_train, y_train, cv=5)
print("Corss validation scores :{}".format(scores))
print("Corss validation means :{:.3f}".format(scores.mean()))
```

```python
prediction = RF_model.predict(X_test)
prediction
```

```python
prediction_rf = gridsearch_rf.predict(X_test)
prediction_rf
```

```python
metrics.accuracy_score(y_test, prediction)
```

```python
get_clf_eval(y_test, prediction_rf)
```
